```python
from bs4 import BeautifulSoup
import requests
```

```python
url="https://www.cars24.com/blog/10-most-expensive-cars-in-the-world/"
```

```python
page=requests.get(url)
```

```python
soup =BeautifulSoup(page.text,'html')
print(soup)
```

```python
print(soup.text)
```

```python
t = soup.find('table')
print(t)
```

```python
data = t.find_all('td')
print(data)
```

```python
data_act = [i.text for i in data]
```

```python
print(data_act)
```

```python
import pandas as pd
```

```python
newdf= pd.DataFrame(columns = ['Model','Body Type','Fuel Type','Ex-showroom Price'])
```

```python
newdf
```

```python
data_act[0:4]
print(len(data_act))
```

```python
ind = 0
st = 0
stop = 4
for i in data_act:
  if stop <= 40:
    newdf.loc[ind]= data_act[st:stop]
    st = st+4
    stop = stop+4
    ind = ind+1
```

```python
newdf
```

```python

newdf.to_csv("CARS.csv")
```
